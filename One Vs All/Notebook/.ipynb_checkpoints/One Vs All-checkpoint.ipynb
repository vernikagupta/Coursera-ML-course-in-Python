```python
# import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.optimize as opt
import scipy.io as sio
from math import sqrt

# load data
data = sio.loadmat('dataset/ex3data1.mat')
x = data['X']
y = data['y']
x.shape, y.shape

# Plot data
_, axarr = plt.subplots(10,10,figsize=(10,10))
for i in range(10):
    for j in range(10):
       axarr[i,j].imshow(x[np.random.randint(x.shape[0])].\
reshape((20,20), order = 'F'))          
       axarr[i,j].axis('off')
        
# insert intercept tem
m = len(x)
ones = np.ones((m,1))
x = np.hstack((ones,x))
x.shape

# sigmoid function
def sigmoid(z):
    return 1/(1+np.exp(-z))
sigmoid(0)

# initialize zero
m,n = x.shape
theta = np.zeros((n,1))
x.shape, theta.shape, y.shape

# define cost function
def cost(theta, x,y,lambd):
    theta_temp = theta[1:]
    J = (-1/m) * np.sum(np.multiply(y, np.log(sigmoid(x @ theta))) 
        + np.multiply((1-y), np.log(1 - sigmoid(x @ theta)))) + ((lambd/2*m)*np.sum(np.power(theta_temp,2)))
    return J 

  
# def gradient function
def gradient(theta, x,y ,lambd):
    m = len(x)
    grad = np.zeros((m,1))
    grad = ((1/m) * x.T @ (sigmoid(x @ theta) - y))
    grad[1:] = grad[1:] + (lambd/m) * (theta[1:])
    return grad 


# train model using optimizing parameters
lambd = 0.1
k = 10
all_theta = np.zeros((k,n))
for i in range(k):
    if i == 0:
        digit_class = 10
    else:
        digit_class = i
    all_theta[i] = opt.fmin_cg(f = cost, x0 = theta,  fprime = gradient, args = (x, (y == digit_class).flatten(), lambd), maxiter = 50)

# Predict accuracy
pred = np.argmax(x @ theta, axis = 1)
pred = [e if e else 10 for e in pred]
np.mean(pred == y.flatten()) * 100
```
